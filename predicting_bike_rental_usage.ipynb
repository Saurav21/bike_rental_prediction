```python
data = data.drop(["instant","casual","registered"], axis = 1)

data.head()
```

```python
data.dteday = pd.to_datetime(data.dteday,dayfirst= True)

data.head()
```

```python
data.set_index("dteday", inplace= True)
```

```python
data.head()
```

```python
data_hot = data[["season", "yr","mnth","holiday","weekday","workingday","weathersit"]]

data_hot.head()
```

```python
# hot coding the data

from sklearn.preprocessing import OneHotEncoder

hot_encoder = OneHotEncoder()
data_hot_encoded = hot_encoder.fit_transform(data_hot).toarray()

```

```python
data_hot_encoded.shape
```

```python
data_hot_encoded = pd.DataFrame(data_hot_encoded)

data_hot_encoded.head()
```

```python
data_cat = data[["temp", "hum","windspeed","cnt"]]

data_cat.head()
```

```python
data_cat.dtypes
```

```python
data_cat = data_cat.reset_index()

data_cat.head()
```

```python
data_cat = data_cat.drop("dteday", axis= 1)

data_cat.head()
```

```python
train_data = pd.concat([data_hot_encoded, data_cat], axis = 1)

train_data.head()
```

```python
x = train_data.iloc[:,:-1].values

x.shape
```

```python
y = train_data.iloc[:,-1:].values

y.shape
```

```python
# scaling the  y data

from sklearn.preprocessing import MinMaxScaler

scalar = MinMaxScaler()

y = scalar.fit_transform(y)

```

```python
# splitting the data

from sklearn.model_selection import train_test_split

x_train,x_test, y_train,y_test = train_test_split(x, y , test_size = 0.2, random_state = 42)

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# model building

import tensorflow as tf
from tensorflow.keras import layers

inputs = layers.Input(shape=(35,))

x = layers.Dense( 100, activation= "relu")(inputs)
x = layers.Dense(100, activation= "relu")(x)
x = layers.Dense(100, activation= "relu")(x)

outputs = layers.Dense(1, activation= "linear")(x)

model = tf.keras.Model(inputs, outputs)

# compiling the model

model.compile(
    loss = tf.keras.losses.mean_squared_error,
    optimizer = tf.keras.optimizers.Adam(0.001)
)


# fitting the model
model_hist = model.fit(x_train,y_train,
                       epochs = 20,
                       validation_data = (x_test,y_test)
                       )




```

```python
model_hist.history.keys()
```

```python
import matplotlib.pyplot as plt

plt.plot(model_hist.history["loss"])
plt.plot(model_hist.history["val_loss"])

plt.show()
```

```python
# model prediction

y_predict = model.predict(x_test)
```

```python
# calculating the various metrics

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

def calculate_metrics(x_test, y_test, y_predict):
  
  k = x_test.shape[1]
  n = len(x_test)

  r_square = r2_score(y_test, y_predict)
  MSE = mean_squared_error(y_test,y_predict)
  MAE = mean_absolute_error(y_test, y_predict)
  adj_r2 = 1-(1-r_square)*(n-1)/(n-k-1)

  result = {
      "R square" : r_square,
      "Adjusted R square": adj_r2,
      "Mean Square Error": MSE,
      "Mean Absolute Error": MAE
  }

  return result
```

```python
print("Metrics : \n", calculate_metrics(x_test,y_test,y_predict))
```
